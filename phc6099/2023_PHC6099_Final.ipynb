#### Background

Bike sharing systems are new generation of traditional bike rentals where whole process from membership, rental and return 
back has become automatic. Through these systems, user is able to easily rent a bike from a particular position and return 
back at another position. Currently, there are about over 500 bike-sharing programs around the world which is composed of 
over 500 thousands bicycles. Today, there exists great interest in these systems due to their important roles in traffic, 
environmental and health issues. 

Apart from interesting real world applications of bike sharing systems, the characteristics of data being generated by
these systems make them attractive for the research. Opposed to other transport services such as bus or subway, the duration
of travel, departure and arrival position is explicitly recorded in these systems. This feature turns bike sharing system into
a virtual sensor network that can be used for sensing mobility in the city. Hence, it is expected that most of important
events in the city could be detected via monitoring these data.

### Question 2. (9 pts)
Write a function, the input is a vector. The output is a **dictionary** of max, min, mean, median, range (i.e., max - min),interquartile range (IQR) of the vector. Then apply this function to the temp attribute of the entire data.

```python
def temp_dict(temp):
    temp = np.array(temp)

    max_value = float(np.max(temp))
    min_value = float(np.min(temp))
    mean_value = float(np.mean(temp))
    median_value = float(np.median(temp))
    range_value = max_value - min_value

    x = np.percentile(temp, 25)
    y = np.percentile(temp, 75)
    iqr = float(y - x)

    dictionary = {
        "max": max_value,
        "min": min_value,
        "mean": mean_value,
        "median": median_value,
        "range": range_value,
        "IQR": iqr
    }

    return dictionary

temp_dict(data["temp"])
```

### Question 3. (9 pts)
Prepare a summary tables by season in year 2011. 
The result should look like the table below.
For age (continuous variable), calculate its mean and standard devition for each group, and keep 2 digits after decimal.
Note that you also need to use the exact row index or column index in the final result. 


| Type            |   spring |  summer | fall | winter |
| --- | --- | --- |  --- |  --- |  
| `N`             |    count       | count |  count |  count |  
| `temp`   |   mean (SD)   | mean (SD)   |  mean (SD)   |  mean (SD)   |  
| `hum`   |   mean (SD)   | mean (SD)   |  mean (SD)   |  mean (SD)   |  
| `windspeed`   |   mean (SD)   | mean (SD)   |  mean (SD)   |  mean (SD)   |  



```python
data_2011 = data[data["yr"] == 0]
cols = ["cnt", "temp", "hum", "windspeed"]
summary = pd.DataFrame(columns=["spring", "summer", "fall", "winter"], index=["N", "temp", "hum", "windspeed"])

for x in cols:
    for season in [1, 2, 3, 4]: 
        season_data = data_2011[data_2011["season"] == season]
        
        if x == "cnt":
            if not season_data.empty:  
                summary.loc["N", ["spring", "summer", "fall", "winter"][season-1]] = season_data[x].sum()
        else:
            if not season_data.empty:  
                mean_val = season_data[x].mean()
                std_val = season_data[x].std()
                summary.loc[x, ["spring", "summer", "fall", "winter"][season-1]] = f"{mean_val:.2f} ({std_val:.2f})"

print(summary)
```

## Question 5. (9 pts)
Draw histograms for count of total rental bikes (cnt), and stratified by seasons using data in 2011. 
The height of the histogram should represent the density instead of the frequency. 
Also draw the density function (in red color) on top of the histogram.



```python
data_2011 = data[data["yr"] == 0]

season_names = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}

plt.figure(figsize=(10, 8))

bins = 20

for season in [1, 2, 3, 4]:
    
    season_data = data_2011[data_2011["season"] == season]
    
    sns.histplot(season_data["cnt"], kde=True, stat="density", bins=bins, label=season_names[season],
                 color=sns.color_palette("Set1")[season-1])

plt.title("Density of Rental Bikes by Season in 2011", fontsize=16)
plt.xlabel("Total Rental Bikes", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.legend(title="Season", loc="upper right")

plt.show()
```
